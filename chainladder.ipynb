# Import libraries

```python
import chainladder as cl
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

# Read data

```python
data_df = pd.read_csv('data.csv')
data_df.head()
```

# Calculate the triangle with incurred loss

```python
tri = cl.Triangle(data=data_df,
            origin='AccidentDate', development='ReportDate',
            columns='Incurred', cumulative=True)
```
